```python
import pandas as pd

class Nodo:
    def __init__(self, no, nom_propietario, vereda_barrio, tipo, sexo, nombre_animal, raza, peso, edad):
        self.no = no
        self.nom_propietario = nom_propietario
        self.vereda_barrio = vereda_barrio
        self.tipo = tipo
        self.sexo = sexo
        self.nombre_animal = nombre_animal
        self.raza = raza
        self.peso = peso
        self.edad = edad
        self.siguiente = None
        self.anterior = None

class ListaDobleEnlazada:
    def __init__(self):
        self.cabeza = None

    def insertar(self, no, nom_propietario, vereda_barrio, tipo, sexo, nombre_animal, raza, peso, edad):
        nuevo_nodo = Nodo(no, nom_propietario, vereda_barrio, tipo, sexo, nombre_animal, raza, peso, edad)
        if not self.cabeza:
            self.cabeza = nuevo_nodo
        else:
            actual = self.cabeza
            while actual.siguiente:
                actual = actual.siguiente
            actual.siguiente = nuevo_nodo
            nuevo_nodo.anterior = actual

    def mostrar_adelante(self):
        if not self.cabeza:
            print("La lista está vacía.")
            return
        actual = self.cabeza
        while actual:
            print(f"No: {actual.no}\nNombre Propietario: {actual.nom_propietario}\nVereda/Barrio: {actual.vereda_barrio}\nTipo: {actual.tipo}\nSexo: {actual.sexo}\nNombre del Animal: {actual.nombre_animal}\nRaza: {actual.raza}\nPeso: {actual.peso}\nEdad: {actual.edad}\n")
            actual = actual.siguiente
        print("None")

    def mostrar_atras(self):
        if not self.cabeza:
            print("La lista está vacía.")
            return
        actual = self.cabeza
        while actual.siguiente:
            actual = actual.siguiente
        while actual:
            print(f"No: {actual.no}\nNombre Propietario: {actual.nom_propietario}\nVereda/Barrio: {actual.vereda_barrio}\nTipo: {actual.tipo}\nSexo: {actual.sexo}\nNombre del Animal: {actual.nombre_animal}\nRaza: {actual.raza}\nPeso: {actual.peso}\nEdad: {actual.edad}\n")
            actual = actual.anterior
        print("None")

    def ordenar(self, atributo):
        if not self.cabeza:
            return

        actual = self.cabeza
        while actual.siguiente:
            siguiente_nodo = actual.siguiente
            while siguiente_nodo:
                if getattr(actual, atributo) > getattr(siguiente_nodo, atributo):
                    actual.no, siguiente_nodo.no = siguiente_nodo.no, actual.no
                    actual.nom_propietario, siguiente_nodo.nom_propietario = siguiente_nodo.nom_propietario, actual.nom_propietario
                    actual.vereda_barrio, siguiente_nodo.vereda_barrio = siguiente_nodo.vereda_barrio, actual.vereda_barrio
                    actual.tipo, siguiente_nodo.tipo = siguiente_nodo.tipo, actual.tipo
                    actual.sexo, siguiente_nodo.sexo = siguiente_nodo.sexo, actual.sexo
                    actual.nombre_animal, siguiente_nodo.nombre_animal = siguiente_nodo.nombre_animal, actual.nombre_animal
                    actual.raza, siguiente_nodo.raza = siguiente_nodo.raza, actual.raza
                    actual.peso, siguiente_nodo.peso = siguiente_nodo.peso, actual.peso
                    actual.edad, siguiente_nodo.edad = siguiente_nodo.edad, actual.edad

                siguiente_nodo = siguiente_nodo.siguiente
            actual = actual.siguiente

df_excel = pd.read_csv('Esterilizacion_canina_y_felina_2021_Puente_Nacional.csv')

lista_doble = ListaDobleEnlazada()

for index, row in df_excel.iterrows():
    lista_doble.insertar(
        row['No.'],
        row['Nom. Propietario'],
        row['Vereda/Barrio'],
        row['Tipo'],
        row['Sexo'],
        row['Nombre del Animal'],
        row['Raza'],
        row['Peso'],
        row['Edad']
    )

print("Lista de adelante hacia atrás:")
lista_doble.mostrar_adelante()
print("SEPARACION_ORDEN_DE_LISTA")


while True:
    respuesta = input("¿Desea insertar más datos? (S/N): ")
    if respuesta.lower() != "s":
        break

    no = input("Ingrese el número: ")
    nom_propietario = input("Ingrese el nombre del propietario: ")
    vereda_barrio = input("Ingrese la vereda/barrio: ")
    tipo = input("Ingrese el tipo: ")
    sexo = input("Ingrese el sexo: ")
    nombre_animal = input("Ingrese el nombre del animal: ")
    raza = input("Ingrese la raza: ")
    peso = float(input("Ingrese el peso: "))
    edad = input("Ingrese la edad: ")

    lista_doble.insertar(no, nom_propietario, vereda_barrio, tipo, sexo, nombre_animal, raza, peso, edad)

print("\nLista ordenada por nombre del propietario:")
lista_doble.ordenar('nom_propietario')
lista_doble.mostrar_adelante()
print("SEPARACION_ORDEN_DE_LISTA")

print("\nLista de atrás hacia adelante:")
lista_doble.mostrar_atras()
print("SEPARACION_ORDEN_DE_LISTA")
```
